```python
# Use this to get rid of the first row 
# Also gets rid of the bottom footers 
df_LoanStatsData = pd.read_csv('LoanStats_2018Q4.csv', 
                 skiprows=1, 
                 skipfooter=2, 
                 engine='python')

# gets rid of columns entirely made of NaN values
df_LoanStatsData = df_LoanStatsData.drop(['url', 'member_id', 'desc', 'id'], axis=1)


print(df_LoanStatsData.shape)
df_LoanStatsData.head()
```

```python
# Modifying string values representing int values into int values


# Use this to change string datatypes of numbers into int values
## ex.) changing "12 %" into .12
# int_rate looks like strings, should be numbers ex: " 14.47%"

# Use .strip() to remove parts of the string and float to turn string into int

int_value = " $23"
#first .strip() removes space, second one removes $
int_value.strip().strip("$")
# This makes '23' which is a string, use float() to make into float
float(int_value.strip().strip("$"))

# Use .strip and float() to create a function

def floatify(values):
  'Takes in a string value and removes ' ' and "%" then returns it as a float'
  return float(values.strip().strip("%"))

# Create a new column called int_rate_float
# Uses the .apply() to apply a function to all values inside column
df_LoanStatsData['int_rate_float'] = df_LoanStatsData['int_rate'].apply(floatify)

df_LoanStatsData.head()
```
